```python
def calc_result(vals):
  a,b = vals
  if(a==b):
    return (3+a,3+b)
  if a==1 and b==3:
    return (6+a,0+b)
  if a==3 and b==1:
    return (0+a,6+b)
  if a>b:
    return (6+a,0+b)
  if a<b:
    return (0+a,6+b)
```

```python
f = open("input", "r")

total_score = (0,0)

for line in f:
  move = (line[0], line[2])
  move_code = (ord(move[0]) - 64, ord(move[1]) - 87)
  score = calc_result(move_code)

  total_score = (total_score[0] + score[0], total_score[1] + score[1])

  # print(f'opponent: {move[0]} ({move_code[0]}) me: {move[1]} ({move_code[1]}) {score}')

print(total_score)
```

```python
def find_move(op_move, result_should_be):
  # X means you need to lose, Y means you need to end the round in a draw, and Z means you need to win.
  if result_should_be == 'Y': # draw
    return op_move
  if result_should_be == 'X': # lose
    if op_move == 1: return 3
    return op_move - 1
  if result_should_be == 'Z':
    if op_move == 3: return 1
    return op_move + 1
```

```python
f = open("input", "r")

total_score = (0,0)
for line in f:
  
  # 1 for Rock, 2 for Paper, and 3 for Scissors
  
  opponent_move = ord(line[0]) - 64
  result_should_be = line[2]

  move_code = (opponent_move, find_move(opponent_move, result_should_be))

  score = calc_result(move_code)

  total_score = (total_score[0] + score[0], total_score[1] + score[1])

  print(f'opponent: {move[0]} ({move_code[0]}) me: {move[1]} ({move_code[1]}) {score}')

print(total_score)
```
