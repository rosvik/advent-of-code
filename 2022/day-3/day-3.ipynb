```python
def prio(ch):
  prio = 0
  if ord(ch) < 96:
    prio = ord(ch) - 64 + 26
  else:
    prio = ord(ch) - 96
  return prio
```

```python
f = open("input", "r")

total = 0

for line in f:
  part1 = line[:(int(len(line)/2))].strip()
  part2 = line[(int(len(line)/2)):].strip()

  dupe = ""
  for ch in part1:
    if (ch in part2): dupe = ch


  pri = prio(dupe)

  print(f'{dupe}, {pri} {part1} {part2}')

  total += pri

print(total)
```

```python
f = open("input", "r")

total = 0

lines = [line.strip() for line in f]

for groupn in range(int(len(lines)/3)):
  group = (lines[(groupn*3):(groupn+1)*3])

  dupe = ""
  for ch in group[0]:
    if (ch in group[1]):
      if (ch in group[2]):
        dupe = ch

  pri = prio(dupe)

  print(f'{dupe}, {pri} {group}')

  total += pri

print(total)
```
