```python
f = open("input", "r")

count = 0

for line in f:
  pair = line.strip()
  a,b = pair.split(',')

  rangeA = a.split('-')
  rangeB = b.split('-')
  rangeA = [int(A) for A in rangeA]
  rangeB = [int(B) for B in rangeB]

  if rangeA[0] <= rangeB[0] and rangeA[1] >= rangeB[1]:
    count += 1
  elif rangeB[0] <= rangeA[0] and rangeB[1] >= rangeA[1]:
    count += 1
  print(f'{count} {rangeA} , {rangeB}')

print(count)
```

```python
f = open("input", "r")

count = 0

for line in f:
  pair = line.strip()
  a,b = pair.split(',')

  rangeA = a.split('-')
  rangeB = b.split('-')
  rangeA = [int(A) for A in rangeA]
  rangeB = [int(B) for B in rangeB]

  mat = False

  for A in rangeA:
    if rangeB[0] <= A <= rangeB[1]:
      mat = True
  for B in rangeB:
    if rangeA[0] <= B <= rangeA[1]:
      mat = True


  print(f'{mat} {line}')

  if mat: count += 1
    
print(count)
```
