```python

def isSafe(line):
    previous = None
    direction = None
    for level in line.split(" "):
        level = int(level)
        if previous != None:
            if direction == None:
                if level > previous:
                    direction = "increasing"
                else:
                    direction = "decreasing"
            if level == previous: 
                return False
            if abs(level - previous) > 3:
                return False
            if direction == "increasing" and level < previous:
                return False
            if direction == "decreasing" and level > previous:
                return False
        previous = level
    return True

def pop(l, i):
    return l[:i] + l[i+1:]

def brute(line):
    for n in range(len(line.split(" "))):
        poppedList = pop(line.split(" "), n)
        if isSafe(" ".join(poppedList)):
            return True
    return False

safeCount = 0

for line in testLevels.split("\n"):
    if isSafe(line):
        safeCount += 1
    else:
        # Life is too short, let's brute force it
        if brute(line):
            safeCount += 1

print("testLevels: " + str(safeCount))

safeCount = 0

for line in inputLevels.split("\n"):
    if isSafe(line):
        safeCount += 1
    else:
        # Life is too short, let's brute force it
        if brute(line):
            safeCount += 1

print("inputLevels: " + str(safeCount))
```
